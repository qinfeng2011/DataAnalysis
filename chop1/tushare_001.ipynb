Signature: ts.month_boxoffice(date=None, retry_count=3, pause=0.001)
Docstring:
获取单月电影票房数据
数据来源：EBOT艺恩票房智库
Parameters
------
    date:日期，默认为上一月，格式YYYY-MM
    retry_count : int, 默认 3
              如遇网络等问题重复执行的次数
    pause : int, 默认 0
             重复请求数据过程中暂停的秒数，防止请求间隔时间太短出现的问题
 return
 -------
    DataFrame 
          Irank         排名
          MovieName     电影名称
          WomIndex      口碑指数
          avgboxoffice  平均票价
          avgshowcount  场均人次
          box_pro       月度占比
          boxoffice     单月票房(万)     
          days          月内天数
          releaseTime   上映日期
